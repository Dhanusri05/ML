```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
# Histogram of 'Age'
plt.hist(data['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age')
plt.show()
```

```python
# Histogram of 'Annual Income (k$)'
plt.hist(data['Annual Income (k$)'], bins=10, edgecolor='black')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Count')
plt.title('Histogram of Annual Income')
plt.show()
```

```python
# Histogram of 'Spending Score (1-100)'
plt.hist(data['Spending Score (1-100)'], bins=10, edgecolor='black')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Count')
plt.title('Histogram of Spending Score')
plt.show()
```

```python
# Scatter plot of 'Age' vs 'Spending Score (1-100)'
plt.scatter(data['Age'], data['Spending Score (1-100)'])
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Scatter plot of Age vs Spending Score')
plt.show()
```

```python
# Scatter plot of 'Annual Income (k$)' vs 'Spending Score (1-100)'
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Scatter plot of Annual Income vs Spending Score')
plt.show()
```

```python
# Skewness of 'Age'
print('Skewness of Age:', data['Age'].skew())

# Skewness of 'Annual Income (k$)'
print('Skewness of Annual Income:', data['Annual Income (k$)'].skew())

# Skewness of 'Spending Score (1-100)'
print('Skewness of Spending Score:', data['Spending Score (1-100)'].skew())

# Statistical description of the data
print(data.describe())
```

```python
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
